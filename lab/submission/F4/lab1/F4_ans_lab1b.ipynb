#Step 12. How many different occupations are in this dataset?

```python
print('No of different occupations:')
users['occupation'].value_counts().count()
```

OR

Solution 2:

```python
print("No of different occupations:")
print(users['occupation'].nunique())
```

OR

Solution 2:

```python
unique_occupation =0
occupation_list=[]

for index in range(0,len(users['occupation'])):
  if users['occupation'].iloc[index] not in occupation_list:
    occupation_list.append(users['occupation'].iloc[index])
    unique_occupation +=1
print ("No of different occupations: \t" + str(unique_occupation))
print ("\n")
unique_occupation_df=pd.DataFrame(occupation_list,index=pd.RangeIndex(start=1,stop=(len(occupation_list)+1)),columns=["unique_occupation"])
print ("List of different occcupation: \n")
print (unique_occupation_df)
```
